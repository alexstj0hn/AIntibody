```python
import pandas as pd
import numpy as np

# Load the CSV file
csv_path = "data/AIntibody_COMPETITION_2_3.csv"  # Replace with the actual path
df = pd.read_csv(csv_path)  # Change separator if needed
df = df.dropna(subset=['cluster_cdr3_heavy', 'cdr1_aa_heavy', 'cdr2_aa_heavy', 'cdr3_aa_heavy', 'cdr1_aa_light', 'cdr2_aa_light', 'cdr3_aa_light', 'sequence_aa_heavy', 'sequence_aa_light'])
df
```

```python
import sys
import threading
import traceback

def dump_all_thread_stacks():
    """Dumps the stack trace of all running threads."""
    print("Dumping stack traces for all threads...\n")
    # Get a dictionary mapping thread identifiers to their current stack frame
    for thread_id, frame in sys._current_frames().items():
        # Find the corresponding thread by matching thread IDs
        thread = next((t for t in threading.enumerate() if t.ident == thread_id), None)
        thread_name = thread.name if thread else "Unknown"

        print(f"--- Thread ID: {thread_id} | Thread Name: {thread_name} ---")
        # Format the stack trace from the frame
        stack_trace = ''.join(traceback.format_stack(frame))
        print(stack_trace)

if __name__ == "__main__":
    dump_all_thread_stacks()
```
