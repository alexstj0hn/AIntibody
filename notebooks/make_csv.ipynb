```python
import pandas as pd
import numpy as np

# Load the CSV file
csv_path = "../data/AIntibody_COMPETITION_2_3.csv"  # Replace with the actual path
df = pd.read_csv(csv_path)  # Change separator if needed
df = df.dropna(subset=['affinity', 'cdr1_aa_heavy', 'cdr2_aa_heavy', 'cdr3_aa_heavy', 'cdr1_aa_light', 'cdr2_aa_light', 'cdr3_aa_light', 'sequence_aa_heavy', 'sequence_aa_light'])
df
```

```python
# Create the concatenated sequence column
df["sequence"] = df.apply(
    lambda row: f"{row['sequence_aa_heavy']}:{row['sequence_aa_light']}",
    axis=1
)
df
```

```python
from sklearn.model_selection import train_test_split

df['binding_affinity'] = 0.001987 * 298 * np.log(df['affinity'])
# standardize the binding affinity
df['binding_affinity'] = (df['binding_affinity'] - df['binding_affinity'].min()) / (df['binding_affinity'].max() - df['binding_affinity'].min())

# Define the number of bins (adjust depending on dataset size)
num_bins = 10  # More bins → finer stratification

# Create bins based on binding affinity distribution
df["affinity_bin"] = pd.qcut(df["binding_affinity"], num_bins, labels=False, duplicates="drop")

train_df, valid_df = train_test_split(df, test_size=0.1, stratify=df["affinity_bin"], random_state=42)

train_df['set'] = 'train'
valid_df['set'] = 'valid'

df = pd.concat([train_df, valid_df])
```

```python
sequences = []
binding_affinities = []
sets = []

for i, row in df.iterrows():    

    vh = [c for c in row['sequence_aa_heavy']]
    vl = [c for c in row['sequence_aa_light']]
    root_sequence = ' '.join(vh) + ' : ' + ' '.join(vl)
    sequences.append(root_sequence)
    binding_affinities.append(row['binding_affinity'])
    sets.append(row['set'])
    
processed_df = pd.DataFrame({
    'sequence': sequences,
    'binding_affinity': binding_affinities,
    'set': sets
})

processed_df['binding_affinity'] = processed_df['binding_affinity'] + np.random.normal(0, 0.0001, processed_df['binding_affinity'].shape)

processed_df.to_csv('../data/comp_2/affinity_sequences.csv', index=False)
```

```python
import pandas as pd
import numpy as np

# Load the CSV file
csv_path = "unlabelled_sequences.csv"  # Replace with the actual path
df = pd.read_csv(csv_path)  # Change separator if needed

# Create the masked sequences with masks between CDRs
sequences = df.apply(
    lambda row: f"{row['sequence_aa_heavy']}:{row['sequence_aa_light']}",
    axis=1
)

sequences = []
masked_sequences = []
for i, row in df.iterrows():

    vh = [c for c in row['sequence_aa_heavy']]
    vl = [c for c in row['sequence_aa_light']]
    root_sequence = ' '.join(vh) + ' : ' + ' '.join(vl)

    sequence = root_sequence

    try:
        for cdr in ['cdr1_aa_heavy', 'cdr2_aa_heavy', 'cdr3_aa_heavy', 'cdr1_aa_light', 'cdr2_aa_light', 'cdr3_aa_light']:
            cdr = ' '.join([c for c in row[cdr]])
            sequence = sequence.replace(cdr, ' '.join(['<mask>'] * len(cdr.split())))
    except:
        continue

    masked_sequences.append(sequence)
    sequences.append(root_sequence)

processed_df = pd.DataFrame({
    'sequence': sequences,
    'masked_sequence': masked_sequences
})

# Save the transformed data
output_path = "processed_unlabelled_sequences.csv"
processed_df.to_csv(output_path, index=False)
processed_df
```

```python
from Bio import pairwise2
from Bio.Seq import Seq
import pandas as pd
import numpy as np
import itertools
from tqdm import tqdm

# Extract sequences (assuming columns are named 'sequence_aa_light' and 'sequence_aa_heavy')
sequences = df['sequence_aa_light'].tolist() + df['sequence_aa_heavy'].tolist()

# Function to compute sequence similarity using pairwise alignment
def sequence_similarity(seq1, seq2):
    alignments = pairwise2.align.globalxx(seq1, seq2, one_alignment_only=True)
    score = alignments[0].score  # Get alignment score (number of matches)
    max_length = max(len(seq1), len(seq2))
    return score / max_length  # Normalize by longest sequence length

# Compute all pairwise similarities
pairwise_similarities = []
for seq1, seq2 in tqdm(itertools.combinations(sequences, 2)):
    similarity = sequence_similarity(seq1, seq2)
    pairwise_similarities.append(similarity)

# Compute average similarity
average_similarity = np.mean(pairwise_similarities)

print(f"Average Sequence Similarity: {average_similarity:.4f}")
```
